```python

def foo_bar(x):
    return np.mean([float(i) for i in x.split('[', 1)[1].split(']')[0].split(',')])

#def foo_bar(x):
#    return print(x)

df_alz_metrics = pd.read_excel('bernn_metrics_all.xlsx', sheet_name='alzheimer metrics values', index_col=0)
df_alz_classif = pd.read_excel('bernn_metrics_all.xlsx', sheet_name='alzheimer classif values', index_col=0)

df_amide_metrics = pd.read_excel('bernn_metrics_all.xlsx', sheet_name='amide metrics values', index_col=0)
df_amide_classif = pd.read_excel('bernn_metrics_all.xlsx', sheet_name='amide classif values', index_col=0)

df_mice_metrics = pd.read_excel('bernn_metrics_all.xlsx', sheet_name='mice metrics values', index_col=0)
df_mice_classif = pd.read_excel('bernn_metrics_all.xlsx', sheet_name='mice classif values', index_col=0)


df_alz_metrics_avg = df_alz_metrics.applymap(foo_bar)
df_alz_classif_avg = df_alz_classif.applymap(foo_bar)
df_amide_metrics_avg = df_amide_metrics.applymap(foo_bar)
df_amide_classif_avg = df_amide_classif.applymap(foo_bar)
df_mice_metrics_avg = df_mice_metrics.applymap(foo_bar)
df_mice_classif_avg = df_mice_classif.applymap(foo_bar)

df_alz_avg = pd.concat((df_alz_classif_avg.loc[:, ['valid/mcc']], df_alz_metrics_avg), 1)
df_amide_avg = pd.concat((df_amide_classif_avg.loc[:, ['valid/mcc']], df_amide_metrics_avg), 1)
df_mice_avg = pd.concat((df_mice_classif_avg.loc[:, ['valid/mcc']], df_mice_metrics_avg), 1)
```

```python
# create a excel writer object
with pd.ExcelWriter("bernn_metrics_new.xlsx") as writer:
    df_alz_metrics.to_excel(writer, sheet_name="alzheimer metrics values")
    df_alz_classif.to_excel(writer, sheet_name="alzheimer classif values")
    df_alz_avg.to_excel(writer, sheet_name="alzheimer avg")
    df_amide_metrics.to_excel(writer, sheet_name="amide metrics values")
    df_amide_classif.to_excel(writer, sheet_name="amide classif values")
    df_amide_avg.to_excel(writer, sheet_name="amide avg")
    df_mice_metrics.to_excel(writer, sheet_name="mice metrics values")
    df_mice_classif.to_excel(writer, sheet_name="mice classif values")
    df_mice_avg.to_excel(writer, sheet_name="mice avg")
```
